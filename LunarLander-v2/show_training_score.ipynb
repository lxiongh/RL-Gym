```python
import json
import numpy as np
from matplotlib import pyplot as plt
from collections import deque


fp = './LunarLander-v2/DQN-lr_1_e-3/openaigym.episode_batch.0.26818.stats.json'

data = json.load(open(fp))

rewards = data['episode_rewards']
plt.plot(range(len(rewards)), rewards, '.c', alpha=0.3)
idx = [x for x in range(len(rewards)) if x%10==0]
r = [rewards[x] for x in range(len(rewards)) if x%10==0]
print(idx[np.argmax(r)])
print(r[np.argmax(r)])


win_deque = deque(maxlen=50)
avg_scores = []

num = len(rewards)
for i in range(num):
    win_deque.append(rewards[i])
    avg_scores.append(np.mean(win_deque))

plt.plot(range(num), avg_scores, '-r')
```
